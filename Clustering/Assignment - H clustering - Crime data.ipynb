```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalizing the data
df_norm = norm_func(crimedata.iloc[:,1:])
df_norm
```

### Second way to normalize 

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(crimedata.iloc[:,1:]))
data 
```

### Normalizing the data to bring all the variables to the same scale and converting the observations into ranges (0-1) to compute the distance.

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch
```

```python
type(df_norm) 
```

```python
z = linkage(df_norm, method="single",metric="euclidean")

plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,
    leaf_font_size=8.,
)
plt.show()
```

```python
z = linkage(df_norm, method="average",metric="euclidean")

plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,
    leaf_font_size=8.,
)
plt.show()
```

```python
z = linkage(df_norm, method="complete",metric="euclidean")

plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,
    leaf_font_size=8.,
    
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5,linkage='single',affinity = "euclidean").fit(df_norm) 
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
crimedata['clust']=cluster_labels
crimedata
```

```python
crimedata.iloc[:,1:].groupby(crimedata.clust).mean()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
crimedata_subset = pd.DataFrame(scaler.fit_transform(crimedata.iloc[:,1:]))
crimedata_subset 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch  
z = linkage(crimedata_subset, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
)
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_norm, method='complete'))
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_norm, method='complete'))
plt.axhline(y=1.2, color='r', linestyle='--')
plt.show()
```

### The x-axis contains the samples and y-axis represents the distance between these samples. The vertical line with maximum distance is the blue line and hence we can decide a threshold of 1.2 and cut the dendrogram

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(crimedata_subset) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
crimedata['clust']=cluster_labels
crimedata
```

### Since we had five clusters, we have five labels in the output i.e. 0 to 4.

### Plotting the clusters to see how actually our data has been clustered

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(crimedata_subset)

plt.scatter(x=crimedata_subset[0], y=crimedata_subset[3], c=cluster.labels_, cmap='rainbow')
```

### The data points are in the form of five clusters in the above scatter plot
