```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
airh_std=pd.DataFrame(std.fit_transform(airh))
airh_std
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(airh_std,method='single'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(airh_std,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model_hier=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

```python
model_hier.fit(airh_std)
```

```python
model_hier.labels_
```

```python
airh["Hier_cluster"]=pd.DataFrame(model_hier.labels_)
airh.head()
```

```python
plt.scatter(airh["Balance"],airh["Qual_miles"],c=airh["Hier_cluster"])
```
